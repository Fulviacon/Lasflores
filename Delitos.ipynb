```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
ruta_archivo = "snic-provincias delitos en arg.csv"
df = pd.read_csv(ruta_archivo)

# Gráfico de barras: cantidad de hechos por tipo de delito
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df,
    x="codigo_delito_snic_nombre",
    y="cantidad_hechos",
    palette="pastel"
)
plt.xticks(rotation=45, ha='right')
plt.title("Cantidad de Hechos por Tipo de Delito")
plt.xlabel("Tipo de Delito")
plt.ylabel("Cantidad de Hechos")
plt.tight_layout()
plt.show()
```

```python
# Gráfico de barras horizontal: tasa de hechos por tipo de delito
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df,
    y="codigo_delito_snic_nombre",
    x="tasa_hechos",
    palette="pastel"
)
plt.title("Tasa de Hechos por Tipo de Delito")
plt.ylabel("Tipo de Delito")
plt.xlabel("Tasa de Hechos")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
ruta_archivo = "snic-provincias delitos en arg.csv"
df = pd.read_csv(ruta_archivo)

# Verifica las primeras filas para asegurarte de que los datos se cargaron correctamente
print(df.head())

# Preparar datos para el gráfico apilado
# Seleccionamos solo las columnas necesarias para el apilamiento
df_apilado = df.set_index("codigo_delito_snic_nombre")[["cantidad_victimas_masc", "cantidad_victimas_fem"]]

# Opcional: Ordenar los delitos por cantidad total de víctimas para una mejor visualización
df_apilado["total_victimas"] = df_apilado["cantidad_victimas_masc"] + df_apilado["cantidad_victimas_fem"]
df_apilado = df_apilado.sort_values("total_victimas", ascending=False).drop("total_victimas", axis=1)

# Crear el gráfico apilado
plt.figure(figsize=(14, 8))
df_apilado.plot(
    kind="bar",
    stacked=True,
    color=["#FFC2C2", "#9EC2FF"],
    figsize=(14, 8)
)

# Personalizar el gráfico
plt.title("Cantidad de Víctimas por Género y Tipo de Delito", fontsize=16)
plt.xlabel("Tipo de Delito", fontsize=14)
plt.ylabel("Cantidad de Víctimas", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.legend(["Víctimas Masculinas", "Víctimas Femeninas"], fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Filtrar datos de una provincia específica, por ejemplo, "Ciudad Autónoma de Buenos Aires"
df_caba = df[df["provincia_nombre"] == "Ciudad Autónoma de Buenos Aires"]

plt.figure(figsize=(10, 6))
sns.barplot(
    data=df_caba,
    x="anio",
    y="cantidad_hechos",
    hue="codigo_delito_snic_nombre",
    palette="pastel"
)
plt.title("Cantidad de Hechos en CABA por Año y Tipo de Delito")
plt.xlabel("Año")
plt.ylabel("Cantidad de Hechos")
plt.legend(title="Tipo de Delito", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
ruta_archivo = "snic-provincias delitos en arg.csv"
df = pd.read_csv(ruta_archivo)

# Verificar las primeras filas para entender los datos
print(df.head())

# Definir los delitos a filtrar
delitos_filtrar = [
    "Robos",
    "Hurtos",
    "Contravenciones",
    "Daños",
    "Acceso ilegal a sistemas informáticos",
    "Daños informáticos",
    "Ciberdelitos sexuales vinculados a menores"
]

# Filtrar los datos para los delitos seleccionados
df_filtrado = df[df["codigo_delito_snic_nombre"].isin(delitos_filtrar)]

# Preparar datos para el gráfico apilado
df_apilado = df_filtrado.set_index("codigo_delito_snic_nombre")[["cantidad_victimas_masc", "cantidad_victimas_fem"]]

# Crear el gráfico apilado
plt.figure(figsize=(14, 8))
df_apilado.plot(
    kind="bar",
    stacked=True,
    color=["#FFC2C2", "#9EC2FF"],
    figsize=(14, 8)
)

# Personalizar el gráfico
plt.title("Cantidad de Víctimas por Género y Tipo de Delito (Delitos Seleccionados)", fontsize=16)
plt.xlabel("Tipo de Delito", fontsize=14)
plt.ylabel("Cantidad de Víctimas", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.legend(["Víctimas Masculinas", "Víctimas Femeninas"], fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtrar los datos
delitos_filtrar = [
    "Robos", "Hurtos", "Contravenciones", "Daños",
    "Acceso ilegal a sistemas informáticos", "Daños informáticos",
    "Ciberdelitos sexuales vinculados a menores"
]
df_filtrado = df[df["codigo_delito_snic_nombre"].isin(delitos_filtrar)]

# Agregar columna de total víctimas y filtrar los principales
df_filtrado["total_victimas"] = df_filtrado["cantidad_victimas_masc"] + df_filtrado["cantidad_victimas_fem"]
df_top = df_filtrado.nlargest(10, "total_victimas")  # Seleccionar los 10 principales

# Preparar los datos
df_apilado = df_top.set_index("codigo_delito_snic_nombre")[["cantidad_victimas_masc", "cantidad_victimas_fem"]]

# Crear el gráfico
plt.figure(figsize=(18, 10))
df_apilado.plot(
    kind="bar",
    stacked=True,
    color=["#FFC2C2", "#9EC2FF"]
)

# Personalizar el gráfico
plt.title("Cantidad de Víctimas por Género y Tipo de Delito (Principales Delitos)", fontsize=16)
plt.xlabel("Tipo de Delito", fontsize=14)
plt.ylabel("Cantidad de Víctimas", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.legend(["Víctimas Masculinas", "Víctimas Femeninas"], fontsize=12)
plt.tight_layout()
plt.show()
```
